# Signal Injection Test

To assess the sensitivity of our search algorithm, we injected artificial laser-like Gaussian spikes into spectral data from HARPS.  We started with single HARPS observations of 20 stars, of spectral types F,G,K and M, at distances between 3 and 170 parsecs.  For each of these, we added an artificial Gaussian spike at a random wavelength, with a FWHM equal to the HARPS spectral resolution at that wavelength, and a peak amplitude corresponding to a random laser power between 500 W and 10 MW, and checked to see if our search algorithm detected it.  We repeated this process 100 times for random laser powers in the bins (500 W-1 kW), (1 kW-10 kW),(10 kW-100 kW),(100 kW-1 MW),(1 MW-10 MW).
